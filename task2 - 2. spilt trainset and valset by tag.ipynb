```python
import json
import pandas as pd
```

# setting.

```python
# Loading data.
train_path = '/content/drive/My Drive/Colab Notebooks/General Model/train.jsonl'
val_path = '/content/drive/My Drive/Colab Notebooks/General Model/val.jsonl'
test_path = '/content/drive/My Drive/Colab Notebooks/General Model/test.jsonl'

def load_jsonl(file_path):
    with open(file_path, 'r') as f:
        return [json.loads(line) for line in f]

train_data = load_jsonl(train_path)
val_data = load_jsonl(val_path)
test_data = load_jsonl(test_path)

# Checking the data.
print(f"Train samples: {len(train_data)}")
print(f"Validation samples: {len(val_data)}")
print(f"Test samples: {len(test_data)}")
```

```python
# Classify the training set and test set according to tag and export csv documents.
def filter_by_tag(data, tag_value):
    return [item for item in data if item.get("tags", [""])[0] == tag_value]

def save_jsonl(data, file_path):
    with open(file_path, 'w') as f:
        for item in data:
            f.write(json.dumps(item) + '\n')

for tag in ["phrase", "passage", "multi"]:
    # 过滤
    train_filtered = filter_by_tag(train_data, tag)
    val_filtered = filter_by_tag(val_data, tag)

    # 保存
    save_jsonl(train_filtered, f"/content/drive/My Drive/Colab Notebooks/train_{tag}.json")
    save_jsonl(val_filtered, f"/content/drive/My Drive/Colab Notebooks/val_{tag}.json")

    # 打印数量确认
    print(f"{tag.capitalize()} - Train: {len(train_filtered)}, Val: {len(val_filtered)}")
```
